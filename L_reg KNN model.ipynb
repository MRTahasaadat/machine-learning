```python
import time
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

samples = [100, 500, 1000, 10000, 100000, 500000]
models = [KNeighborsClassifier(n_neighbors=3), LogisticRegression()]
Data = []

# Collecting data for each model and sample size
for model in models:
    for n_samples in samples:
        X, y = make_blobs(
            n_samples=n_samples,
            n_features=2,
            centers=10,
            cluster_std=9.7,
            random_state=10
        )

        x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=10)

        print(model.__class__.__name__, " ...", end=" ")

        t0 = time.time()
        model.fit(x_train, y_train)
        train_time = time.time() - t0

        t0 = time.time()
        model_pred = model.predict(x_test)
        predict_time = time.time() - t0

        accuracy = accuracy_score(y_test, model_pred)
        f1 = f1_score(y_test, model_pred, average="weighted")
        precision = precision_score(y_test, model_pred, average="weighted")
        recall = recall_score(y_test, model_pred, average="weighted")

        # Saving the results
        data_dict = {
            "model": model.__class__.__name__,
            "n_samples": n_samples,
            "train_time": train_time,
            "predict_time": predict_time,
            "accuracy": accuracy,
            "f1": f1,
            "precision": precision,
            "recall": recall
        }

        Data.append(data_dict)
        print("Done")

# Creating a DataFrame from the collected data
df = pd.DataFrame(Data)

# Filtering data for KNN and Logistic Regression
df_knn = df[df['model'] == 'KNeighborsClassifier']
df_lr = df[df['model'] == 'LogisticRegression']

# Plotting the results
plt.figure(figsize=(12, 6))

# Plotting time (training and prediction)
plt.subplot(2, 1, 1)
plt.title("Model Training and Prediction Time")
plt.plot(df_knn['n_samples'], df_knn['train_time'], "b-", label="KNN Train Time")
plt.plot(df_lr['n_samples'], df_lr['train_time'], "g-", label="LR Train Time")
plt.plot(df_knn['n_samples'], df_knn['predict_time'], "b--", label="KNN Predict Time")
plt.plot(df_lr['n_samples'], df_lr['predict_time'], "g--", label="LR Predict Time")
plt.xlabel("Number of Samples")
plt.ylabel("Time (seconds)")
plt.legend()

# Plotting F1 scores
plt.subplot(2, 1, 2)
plt.title("Model F1 Score Comparison")
plt.plot(df_knn['n_samples'], df_knn['f1'], "b-", label="KNN F1 Score")
plt.plot(df_lr['n_samples'], df_lr['f1'], "g-", label="LR F1 Score")
plt.xlabel("Number of Samples")
plt.ylabel("F1 Score")
plt.legend()

plt.tight_layout()
plt.show()
```
